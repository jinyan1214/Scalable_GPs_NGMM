```python
import geopandas as gpd
import sys
sys.path.append("/resnick/groups/enceladus/jyzhao/Scalable_GPs_jz/regression_gpytorch/src")
from training_utils.utils import read_hdf5_with_site_id
```

```python
num_training_eq = 400
X, Y, site_id = read_hdf5_with_site_id(f'/resnick/groups/enceladus/jyzhao/Scalable_GPs_jz/regression_gpytorch/output/formatted_data/training_{num_training_eq}_eqs/training_sites_training_eqs_2.00_ASK14_{num_training_eq}_eqs_1_var_per_eq.h5')
gdf = gpd.GeoDataFrame(
    {'site_id': site_id, 'eq_rup_id': X[:, 0], 'Y': Y,
     'mpt_x': X[:, 4]*1000, 'mpt_y': X[:, 5]*1000},
    geometry=gpd.points_from_xy(X[:, 4]*1000, X[:, 5]*1000),
    crs="EPSG:32611"
)
```

```python
(gdf.groupby('eq_rup_id').nunique()['geometry'] > 1).any()
```

```python
unique_eq_rup = gdf[['eq_rup_id', 'mpt_x', 'mpt_y', 'geometry']].drop_duplicates().sort_values('eq_rup_id')
unique_eq_rup.explore()
```
