```python
import requests

# Define the URL you want to send the POST request to
url = 'https://nsi.sec.usace.army.mil/nsiapi/structures?fmt=fc'

# Define the data you want to send (typically in JSON format)
data = {
	"type": "FeatureCollection",
	"features": [{
		"type": "Feature",
		"geometry": {
			"type": "Polygon",
			"coordinates": [
				[
					[-121.504025, 38.588017],
					[-121.507201, 38.596202],
					[-121.503081, 38.599623],
					[-121.494153, 38.601568],
					[-121.484624, 38.600428],
					[-121.478701, 38.597342],
					[-121.473292, 38.590835],
					[-121.504025, 38.588017]
				]
			]
		},
		"properties": {}
	}]
}

# Send the POST request with the data
response = requests.post(url, json=data)

# Check the status code and response
if response.status_code == 200:
    print('Request was successful!')
    print(response.json())  # If the response is JSON
else:
    print(f'Failed to make request: {response.status_code}')
```
